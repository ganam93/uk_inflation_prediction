```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.random.rand(30)
y = np.random.rand(30)

# Create a scatter plot
plt.scatter(x, y, color='green', marker='o')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

# Show the plot
plt.show()
```
